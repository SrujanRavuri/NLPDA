```python
import nltk
from nltk.corpus import brown
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('brown')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
import pandas as pd
from nltk.tokenize import TreebankWordTokenizer
```

```python
from nltk import word_tokenize
tokens = word_tokenize(brown.raw())
tokens
```

#Categories available

```python
brown.categories()
```

#Size of the brown corpus

```python
print("Size of brown corpus:",len(tokens))
```

#Size of word tokens

```python
from nltk.probability import FreqDist
fdist = FreqDist(brown.words())
fdist
```

# Size of category "government"

```python
GovernmentTokenList=brown.words(categories='government')
print(len(GovernmentTokenList))
```

#Size of word types

```python
tags=nltk.pos_tag(brown.words())
from collections import Counter
counts=Counter( tag for word,tag in tags)
print(counts)
```

#Most frequent tokens

```python
Tokencount=nltk.FreqDist(tokens)
print(Tokencount.most_common(50))
```

#No of sentences

```python
print(len(brown.sents()))
```
